Q1. What is an ensemble technique in machine learning?

Ensemble tenchique in machine learning combine multiple models to improve predictive accuracy.This includes methods like bagging , boosting , random forest and stacking.

Q2. Why are ensemble techniques used in machine learning?

Ensemble tenhnique in machine learning are used to improve predictive accuracy , used to reduce overfitting and enhance roburstness against outlier and noise.

Q3. What is bagging?

Bagging is a technique where multiple models are trained on different subsets of the data, generated by random sampling with replacement. The final prediction is typically an aggregation of the predictions from these models.

Q4. What is boosting?

Boosting is a machine learning technique where weak leaners are combined sequentially, each correcting the error of it's  predecessors, to create a strong learner.







Q5. What are the benefits of using ensemble techniques?

Ensemble technique often improve predictive accuracy , reduce overfitting and enhance robustness against noise and outliers and increased stability in predictions.

Q6. Are ensemble techniques always better than individual models?


Ensemble techniques are not always superior to individual models. Factors such as computational resources, interpretability, training time, data quality, and model diversity can influence whether using an ensemble is appropriate.

To calculate confidence interval using bootstrap

. Sample the dataset with replacement to create multiple bootstrap samples.

. Calculate the statistic of interest for each bootstrap sample.

. Determine the desired percentiles (e.g., 2.5th and 97.5th) of the bootstrap sample statistics.

. Define the confidence interval using these percentiles.

Q8. How does bootstrap work and What are the steps involved in bootstrap?

Sample with Replacement: Randomly pick samples from your dataset, allowing duplicates (replacement).

Calculate Statistic: Compute the statistic of interest (like mean or median) for each bootstrap sample.

Repeat: Do this process many times (typically thousands) to create a distribution of your statistic.

Estimate Confidence Interval: Use this distribution to estimate the confidence interval for your statistic.
